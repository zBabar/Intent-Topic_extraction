```python
import pandas as pd
import numpy as np
from utils import Dataset, Preprocessing
from sklearn.cluster import DBSCAN
```

```python
data_obj = Dataset() 

# get labelled data
train_df = data_obj.get_labelled_data()
process_obj = Preprocessing()

cleaned_train_df = process_obj.get_preprocessd_data(train_df, text_column="phrase", has_labels = True)
X, y = process_obj.get_DTM(cleaned_train_df, has_labels = True)

# get unlabelled data
test_df = data_obj.get_unlabelled_data()

cleaned_test_df = process_obj.get_preprocessd_data(test_df, text_column="user_msg", has_labels = False)


X_test = process_obj.get_DTM(cleaned_test_df, has_labels = False)

```

```python
dbscan = DBSCAN(eps=0.5, min_samples=3, metric='cosine')
dbscan.fit(X_test)
```

```python
labels = dbscan.labels_
print("Cluster labels for each document:")
for doc, label in zip(cleaned_test_df['cleaned_text'].to_list(), labels):
    print(f"Document: {doc} => Cluster: {label}")

len(np.unique(labels))
```

```python
from sklearn.cluster import KMeans

true_k = 5  # Number of clusters
model = KMeans(n_clusters=true_k, init='k-means++', max_iter=100, n_init=1)
model.fit(X_test)

# Print top terms per cluster
print("Top terms per cluster:")
order_centroids = model.cluster_centers_.argsort()[:, ::-1]
terms = process_obj.get_vectorizer().get_feature_names_out()
```

```python

for i in range(true_k):
    print(f"Cluster {i}:")
    for ind in order_centroids[i, :10]:
        print(f' {terms[ind]}')
    print()

# Print cluster assignments for each document
labels = model.labels_
print("Cluster assignments for each document:")
for doc, label in zip(cleaned_test_df["user_msg"], labels):
    print(f"Document: {doc} => Cluster: {label}")
```
